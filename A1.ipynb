```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (replace with your file or use a sample)
# Example: df = pd.read_csv("data.csv")
df = sns.load_dataset('iris')  # Sample dataset

# Select numeric columns
numeric_cols = df.select_dtypes(include='number').columns

# Histograms
df[numeric_cols].hist(figsize=(10, 6), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("ðŸ“Š Histograms of Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(10, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, (len(numeric_cols)+1)//2, i)
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (replace with your own if needed)
# df = pd.read_csv('data.csv')
df = sns.load_dataset('iris')  # Sample dataset

# --- Correlation Matrix ---
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("ðŸ”— Correlation Matrix of Numeric Features")
plt.show()

# --- Pairplot ---
sns.pairplot(df, hue=df.columns[-1])  # Hue set to last column (e.g. species)
plt.suptitle("ðŸ” Pairplot of Feature Relationships", y=1.02)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
# df = pd.read_csv('data.csv')
df = sns.load_dataset('iris')  # Sample dataset

numeric_cols = df.select_dtypes(include='number').columns

# --- Boxplots for Outliers ---
plt.figure(figsize=(10, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, (len(numeric_cols)+1)//2, i)
    sns.boxplot(y=df[col], color='tomato')
    plt.title(f'Outliers in {col}')
plt.tight_layout()
plt.show()

# --- Pairplot to Visualize Clusters / Separation ---
sns.pairplot(df, hue=df.columns[-1])
plt.suptitle("Feature Relationships & Potential Clusters", y=1.02)
plt.show()

# --- Correlation Matrix for Trends ---
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Trends Between Features")
plt.show()
```
